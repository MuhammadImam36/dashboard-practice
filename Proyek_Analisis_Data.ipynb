## Exploratory Data Analysis (EDA)

### Merging tabel untuk mendapatkan informasi yang lebih komprehensif

```python
new_df = pd.merge(
    left=orders,
    right=customers,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
new_df.head()
```

```python
new_df2 = pd.merge(
    left=new_df,
    right=order_items,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
new_df2.head()
```

```python
new_df3 = pd.merge(
    left=new_df2,
    right=order_reviews,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
new_df3.head()
```

```python
new_df4 = pd.merge(
    left=new_df3,
    right=products,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
new_df4.head()
```

```python
ecommerce = pd.merge(
    left=new_df4,
    right=product_category,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
ecommerce.head()
```

### Drop kolom yang tidak diperlukan

```python
ecommerce.drop(['customer_id', 'order_status', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'customer_unique_id', 'customer_zip_code_prefix'], axis=1, inplace=True)
```

```python
ecommerce.head()
```

```python
ecommerce.drop(['order_item_id', 'product_id', 'seller_id', 'shipping_limit_date', 'review_answer_timestamp', 'product_name_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'product_description_lenght', 'review_creation_date', 'review_id'], axis=1, inplace=True)
```

### Drop instance yang terdapat null value

```python
ecommerce.dropna(inplace = True)
```

### Tingkat kepuasan kustomer terhadap produk yang dibeli

```python
review_counts = ecommerce['review_score'].value_counts().reset_index()
review_counts.columns = ['review_score', 'frequency']
review_counts
```

### Produk yang paling banyak dibeli oleh kustomer

```python
category_counts = ecommerce.groupby(['product_category_name', 'product_category_name_english']).size().reset_index(name='count')
category_counts = category_counts.sort_values(by='count', ascending=False).reset_index()
```

```python
category_counts[:5]
```

```python
category_counts[-5:]
```

### Produk yang mempunyai revenue terbesar

```python
revenue_counts = ecommerce.groupby(by=['product_category_name', 'product_category_name_english']).agg({
    'price' : 'sum'
})
revenue_counts = revenue_counts.sort_values(by = 'price', ascending = False).reset_index()
```

```python
revenue_counts[:5]
```

```python
revenue_counts[-5:]
```

### Total terjadinya transaksi per bulan

```python
monthly_orders = ecommerce.groupby(ecommerce['order_purchase_timestamp'].dt.strftime('%B')).agg({
    "order_id": "nunique",
}).reset_index()

monthly_orders.rename(columns={
    "order_id": "order_count",
    "order_purchase_timestamp": "month",
}, inplace=True)

month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
monthly_orders['month'] = pd.Categorical(monthly_orders['month'], categories=month_order, ordered=True)
monthly_orders = monthly_orders.sort_values('month')
monthly_orders = monthly_orders.reset_index(drop = True)
monthly_orders
```

**Insight:**
- Kostumer mayoritas puas terhadap produk yang telah dibeli dengan memberikan rating 5/5
  
- Kategori produk yang paling banyak dibeli oleh kustomer yaitu pada kategori bed_bath_table
- Produk yang mempunyai revenue terbesar yaitu beleza_saude pada kategori health_beauty
- Pada bulan agustus, kustomer paling banyak melakukan pembelian di ecommerce tersebut
